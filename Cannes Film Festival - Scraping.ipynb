# I'll try to get the information about Cannes Films Festivals award from Wikipedia


```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
pd.set_option('display.max_rows', None)
```

```python
#Extracting Palm D'Or
```

```python
# Send a request to the Wikipedia page
url = "https://es.wikipedia.org/wiki/Palma_de_Oro"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table with the Palm d'or awards
table = soup.select("#mw-content-text > div.mw-parser-output > table:nth-child(19) > tbody")[0]

# Extract the years, movie names, and directors from the table
years = []
movie_names = []
directors = []
for row in table.find_all("tr")[1:]:  # Exclude the header row
    cells = row.find_all("td")
    year_cell = cells[0].find("a")
    movie_name_cell = cells[1].find("i")
    director_cell = cells[2].find("a")
    if year_cell and movie_name_cell and director_cell:  # Check if the cells contain expected elements
        year = year_cell.text
        movie_name = movie_name_cell.text
        director = director_cell.text
        years.append(year)
        movie_names.append(movie_name)
        directors.append(director)

# Create a dataframe with the extracted data
df1 = pd.DataFrame({"year": years, "movie_name": movie_names, "awarded": directors})

# Print the dataframe
df1
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Send a request to the Wikipedia page
url = "https://es.wikipedia.org/wiki/Anexo:Premio_del_Festival_de_Cannes_a_la_mejor_actriz"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table with the data
table = soup.select("table.sortable.wikitable")[0]

# Extract the years, movie names, directors, and awarded from the table
years = []
movie_names = []
awarded = []
for row in table.find_all("tr")[1:]:  # Exclude the header row
    cells = row.find_all("td")
    if len(cells) >= 4:  # Check if the row has enough cells
        year = cells[0].find("a").text if cells[0].find("a") else None
        movie_name = cells[1].find("i").find("b").find("a").text if cells[1].find("i") and cells[1].find("i").find("b") and cells[1].find("i").find("b").find("a") else None
        awarded_text = cells[3].text
        years.append(year)
        movie_names.append(movie_name)
        awarded.append(awarded_text)


# Create a dataframe with the extracted data
df2 = pd.DataFrame({"year": years, "movie_name": movie_names, "awarded": awarded})

# Print the dataframe
df2
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a request to the Wikipedia page
url = "https://es.wikipedia.org/wiki/Anexo:Premio_del_Festival_de_Cannes_al_mejor_actor"
response = requests.get(url)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find the table with the data
table = soup.select("table.sortable.wikitable")[0]

# Extract the years, movie names, and awarded information from the table
years = []
movie_names = []
awarded = []
for row in table.find_all("tr")[1:]:  # Exclude the header row
    cells = row.find_all("td")
    if len(cells) >= 3:  # Check if the row has enough cells
        year = cells[0].find("a").text if cells[0].find("a") else None
        movie_name = cells[1].find("i").find("b").find("a").text if cells[1].find("i") and cells[1].find("i").find("b") and cells[1].find("i").find("b").find("a") else None
        awarded_text = cells[2].text
        years.append(year)
        movie_names.append(movie_name)
        awarded.append(awarded_text)

# Create a dataframe with the extracted data
df3 = pd.DataFrame({"year": years, "movie_name": movie_names, "awarded": awarded})

# Print the dataframe
df3
```

```python
# Convert 'year' column to integer data type
df1['year'] = df1['year'].astype(int)
df2['year'] = df2['year'].astype(int)
df3['year'] = df3['year'].astype(int)

```

```python
#cleaning df1
```

```python
# Function to extract text inside brackets
def extract_text_inside_brackets(text):
    start_index = text.find("(")
    end_index = text.find(")")
    if start_index != -1 and end_index != -1:
        return text[start_index + 1:end_index]
    else:
        return text

# Apply the function to the 'movie_name' column of the DataFrame
df1['movie_name'] = df1['movie_name'].apply(extract_text_inside_brackets)


df1
```

```python
# Function to filter out rows with text containing square brackets
def filter_out_square_brackets(text):
    if '[' in text and ']' in text:
        return False
    else:
        return True

# Apply the function to the 'awarded' column of the DataFrame
df1 = df1[df1['awarded'].apply(filter_out_square_brackets)]

# Reset the index of the DataFrame
df1 = df1.reset_index(drop=True)
df1
```

```python
print(df1.head())
print(df2.head())
print(df3.head())
```

```python
#cleaning df2

# Use the replace() method to remove '/n' from the end of text
df2['awarded'] = df2['awarded'].str.replace(r'\n$', '', regex=True)

# Display the updated DataFrame
df2
```

```python
#cleaning df3
```

```python
# Use the replace() method to remove '/n' from the end of text and to remove brackets and content inside of them
df3['awarded'] = df3['awarded'].str.replace(r'\n$', '', regex=True)
df3['awarded'] = df3['awarded'].str.replace(r'\s*\(.*\)', '', regex=True)

df3
```

```python
# Function to filter out rows with empty values
def filter_out_empty(text):
    if text.strip() == '':
        return False
    else:
        return True

# Apply the function 
df3 = df3[df3['awarded'].apply(filter_out_empty)]

# Reset the index of the DataFrame
df3 = df3.reset_index(drop=True)

# Display the updated DataFrame
df3
```

```python
# Now, before merge them I want to create a column for each dataframe with the name of the category

# Function to create a new column with a specified value for each row
def create_new_column(df, column_name, value):
    df[column_name] = value
    return df
# Call the function to create the new columns 
df1 = create_new_column(df1, 'cannes_won_category1', 'Palm d\'Or')
df2 = create_new_column(df2, 'cannes_won_category2', 'Best Actress')
df3 = create_new_column(df3, 'cannes_won_category3', 'Best Actor')

# Display the updated DataFrames
print(df1.head())
print(df2.head())
print(df3.head())
```

```python
#Concatenate all
```

```python
import pandas as pd

# Concatenate the three dataframes into one, using the 'merge' method from pandas
# and specifying the common columns 'year' and 'movie_name' as the merge keys
a_cannes = pd.merge(df1, df2, on=['year', 'movie_name'], how='outer')
a_cannes = pd.merge(a_cannes, df3, on=['year', 'movie_name'], how='outer')

# Drop the 'awarded' column from the resulting DataFrame
a_cannes = a_cannes.drop(['awarded','awarded_x', 'awarded_y'], axis=1)

# Create a new column 'cannes_won_num' that counts the values in columns 'cannes_won_category1', 'cannes_won_category2', and 'cannes_won_category3'
a_cannes['cannes_won_num'] = a_cannes[['cannes_won_category1', 'cannes_won_category2', 'cannes_won_category3']].count(axis=1)

# Sort the final dataframe by the 'year' column
a_cannes = a_cannes.sort_values('year')

# Show the resulting DataFrame
a_cannes
```
