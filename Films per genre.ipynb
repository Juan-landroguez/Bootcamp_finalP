# After the first exploration I'm going to start cleanning

```python
## Divide column 'director' in different columns with single director names
```

```python
# Split 'director' column by ", \n" and get the first part
all_films['director'] = all_films['director'].str.split(", \n").str.get(0)

# Verify the updated 'director' column
all_films['director'].head()
```

```python
all_films['director']
```

```python
## Divide column 'Stars' in different columns with single actor names
```

```python
# Split 'star' column into separate columns for each actor
actor_columns = all_films['star'].str.split(", \n", expand=True)

# Determine the maximum number of actors in any row
max_actors = actor_columns.notna().sum(axis=1).max()

# Create new column names based on the maximum number of actors
new_columns = [f'actor{i}' for i in range(1, max_actors + 1)]

# Add new columns to the dataframe
all_films[new_columns] = actor_columns

# Drop the original 'star' column
all_films.drop(columns=['star'], inplace=True)

# Verify the updated dataframe
all_films.head()
```

```python
## Divide column 'genre' in different columns with single genres
```

```python
# Split 'genre' column into separate columns for each genre
genre_columns = all_films['genre'].str.split(",", expand=True)

# Determine the maximum number of genres in any row
max_genre = genre_columns.notna().sum(axis=1).max()

# Create new column names based on the maximum number of genres
new_columns = [f'genre{i}' for i in range(1, max_genre + 1)]

# Add new columns to the dataframe
all_films[new_columns] = genre_columns

# Drop the original 'genre' column
all_films.drop(columns=['genre'], inplace=True)

# Verify the updated dataframe
all_films.head()
```

```python
#Convert runtime into an int
```

```python
# Remove characters after numerical value in 'runtime' column
all_films['runtime'] = all_films['runtime'].str.extract(r'(\d+)', expand=False)
all_films['runtime'] = all_films['runtime'].fillna(0)

# Convert 'runtime' column to integer data type
all_films['runtime'] = all_films['runtime'].astype(int)

# Verify the updated dataframe
all_films.head()
```

```python
# Convert 'votes' column to integer data type
```

```python
all_films['votes'] = all_films['votes'].fillna(0)
all_films['votes'] = all_films['votes'].astype(int)

# Verify the updated dataframe
all_films.head()
```

```python
# Convert 'year' column to integer data type
```

```python
# Replace non-numeric values with 0 in 'year' column
all_films['year'] = all_films['year'].replace(r'\D', 0, regex=True)

# Convert 'year' column to numeric data type
all_films['year'] = pd.to_numeric(all_films['year'], errors='coerce')

# Set values greater than 2023 to 0
all_films.loc[all_films['year'] > 2023, 'year'] = 0

# Convert 'year' column to integer data type
all_films['year'] = all_films['year'].fillna(0).astype(int)

# Drop rows with 0 value in the 'year' column
all_films = all_films[all_films['year'] != 0]


# Verify the unique values updated of 'year' column
print(all_films['year'].unique())

```

```python
all_films.shape
```

```python
# Verify the types after some changes 
all_films.dtypes
```

```python
# Filter the duplicate values based on 'movie_name' column
duplicated_movies = all_films[all_films.duplicated(subset='movie_name', keep=False)]

# Sort the duplicated rows alphabetically by 'movie_name' column
duplicated_movies = duplicated_movies.sort_values('movie_name')

# verify changes
duplicated_movies.head(60)
```

```python
# Count the number of NaN values in each row and store it in a new column 'nan_count'
all_films['nan_count'] = all_films.isna().sum(axis=1)

# Sort the DataFrame by 'nan_count' column in ascending order
all_films = all_films.sort_values('nan_count', ascending=True)

# Drop duplicates based on 'movie_id' column, keeping the row with the least number of NaN values in other columns
all_films = all_films.drop_duplicates(subset='movie_id', keep='first')

# Drop duplicates based on 'movie_name' column, keeping the row with the least number of NaN values in other columns
all_films= all_films.drop_duplicates(subset='movie_name', keep='first')

# Drop the 'nan_count' column as it was only used for sorting
all_films = all_films.drop('nan_count', axis=1)

# Verify changes
all_films.head(60)
```

```python
# Get the count of occurrences for each unique value in the 'certificate' column
certificate_counts = all_films['certificate'].value_counts()

# Print the count of occurrences for each unique value
print(certificate_counts)
```

```python
# Create a dictionary mapping to homogenize the certificates, using MPAA standards
certificate_mapping = {
    'Not Rated': 'Unrated',
    'R': 'NC-17',
    'Approved': 'G',
    'Passed': 'G',
    'PG-13': 'PG-13',
    'PG': 'PG-13',
    'Unrated': 'Unrated',
    'TV-MA': 'NC-17',
    'TV-14': 'PG-13',
    'G': 'G',
    'TV-PG': 'G',
    'TV-G': 'G',
    'GP': 'PG-13',
    'X': 'NC-17',
    'U': 'G',
    'TV-Y7': 'PG-13',
    'M': 'PG-13',
    'UA': 'G',
    'TV-Y': 'PG-13',
    'M/PG': 'PG-13',
    'NC-17': 'NC-17',
    '18+': 'NC-17',
    '13+': 'PG-13',
    'TV-Y7-FV': 'PG-13',
    '16+': 'NC-17',
    'A': 'G',
    'E10+': 'PG-13',
    'Open': 'Unrated',
    'E': 'G',
    '12': 'PG-13',
    'T': 'G',
    'TV-13': 'NC-17',
    'AO': 'NC-17',
    '16': 'NC-17',
    '18': 'NC-17',
    'UA 16+': 'NC-17',
    'UA 7+': 'G',
    'MA-17': 'NC-17',
    '13': 'PG-13',
    'UA 13+': 'PG-13',
    'F': 'Unrated',
    '7': 'G',
    'MA-13': 'PG-13'
}

# filling NaNs with 'Unrated'
all_films['certificate'].fillna('Unrated', inplace=True)

# Apply the mapping
all_films['certificate'] = all_films['certificate'].map(certificate_mapping)

# Verify changes
all_films['certificate'].value_counts()
```

```python
# After understand that most of films are not rated I prefer to drop the column, I'll advise to check certificates after recommend if there is any restriction
all_films = all_films.drop('certificate', axis=1)

# As Gross Revenue column it's mostly empty I drop it
all_films = all_films.drop('gross(in $)', axis=1)
```

```python
#In order to be self explainatory I rename year

all_films.rename(columns={'year': 'year_film'}, inplace=True)
```
