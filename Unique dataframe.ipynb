# Open dataframes

```python
import pandas as pd
import numpy as np
```

```python
all_films = pd.read_csv('all_films_cleaned.csv')
```

# Merge both Dataframes

```python
#Merge the two DataFrames on the 'movie_name' column using left to include all rows from 'all_films'
merged_df = all_films.merge(a_oscars[['movie_name', 'osc_nominated_num', 'osc_won_num',
       'osc_nominated_category1', 'osc_nominated_category2',
       'osc_nominated_category3', 'osc_nominated_category4',
       'osc_nominated_category5', 'osc_nominated_category6',
       'osc_nominated_category7', 'osc_nominated_category8',
       'osc_nominated_category9', 'osc_nominated_category10',
       'osc_nominated_category11', 'osc_nominated_category12',
       'osc_nominated_category13', 'osc_nominated_category14',
       'osc_nominated_category15', 'osc_nominated_category16',
       'osc_nominated_category17', 'osc_nominated_category18',
       'osc_nominated_category19', 'osc_nominated_category20',
       'osc_nominated_category21', 'osc_nominated_category22',
       'osc_won_category1', 'osc_won_category2', 'osc_won_category3',
       'osc_won_category4', 'osc_won_category5', 'osc_won_category6',
       'osc_won_category7', 'osc_won_category8', 'osc_won_category9',
       'osc_won_category10', 'osc_won_category11', 'osc_won_category12']],
                            on='movie_name', how='left')


# You can also use 'how='left'' or 'how='right'' if you want to include all rows from 'all_films' or 'a_oscars' respectively,
# even if there is no match in the other DataFrame

# Now, merged_df contains all the columns from all_films, as well as the selected columns from a_oscars,
# only for the rows where 'movie_name' is the same in both DataFrames

# You can rename merged_df to 'all_films_a' if needed:
films_w_oscars = merged_df.copy()
```

```python
#Merge the two DataFrames on the 'movie_name' column using left to include all rows from 'all_films'
merged_df2 = films_w_oscars.merge(a_cannes[['movie_name', 'cannes_won_category1', 'cannes_won_category2',
       'cannes_won_category3', 'cannes_won_num']],
                            on='movie_name', how='left')


# You can also use 'how='left'' or 'how='right'' if you want to include all rows from 'all_films' or 'a_oscars' respectively,
# even if there is no match in the other DataFrame

# Now, merged_df contains all the columns from all_films, as well as the selected columns from a_oscars,
# only for the rows where 'movie_name' is the same in both DataFrames

# Sort the final dataframe by the 'year' column
merged_df2 = merged_df2.sort_values('year_film')

# Reset the index of the DataFrame
merged_df2 = merged_df2.reset_index(drop=True)

# You can rename merged_df to 'all_films_a' if needed:
films_w_awards = merged_df2.copy()
```

# Some cleaning

```python
# Fill the NaNs of awards with 0 and covert in int

films_w_awards['osc_nominated_num'] = films_w_awards['osc_nominated_num'].fillna(0)
films_w_awards['osc_nominated_num'] = films_w_awards['osc_nominated_num'].astype(int)

films_w_awards['osc_won_num'] = films_w_awards['osc_won_num'].fillna(0)
films_w_awards['osc_won_num'] = films_w_awards['osc_won_num'].astype(int)

films_w_awards['cannes_won_num'] = films_w_awards['cannes_won_num'].fillna(0)
films_w_awards['cannes_won_num'] = films_w_awards['cannes_won_num'].astype(int)


print(films_w_awards['osc_nominated_num'].value_counts())
print(films_w_awards['osc_won_num'].value_counts())
print(films_w_awards['cannes_won_num'].value_counts())
```

```python
# I've identify in descriptions a common text that it's like NaN, I prefer to change it to clear NaN
films_w_awards['description'] = films_w_awards['description'].replace('Add a Plot', '')

films_w_awards.head(60)
```
